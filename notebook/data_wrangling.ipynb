```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/cleaned_car_data.csv')
df.head()
```

```python
df.dtypes
```

```python
numeric_feature = df.select_dtypes(include=[np.number]).columns
print("The skewness of the numerical features are:\n\n", df[numeric_feature].skew().sort_values(ascending=False))
```

##### This data's skewness values are very good and do not require any adjustments.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaler = df.copy()

for column in df_scaler.select_dtypes(exclude='object').columns:
    if column == 'Price':
        continue;
    df_scaler[column] = scaler.fit_transform(df_scaler[[column]])
```

```python
df_scaler.head()
```

```python
df_scaler.to_csv('../data/scaled_data.csv', index=False)
```
